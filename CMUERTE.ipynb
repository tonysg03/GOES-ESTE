# Tiempo

Las fechas aparecen de tipo object, significando que no tiene ningún formato especial, es solo cualquier objeto. Por otro lado las demás variables si son de formato float64. Estas variables se puede utilizar en ese formato por el resto del programa pero las fechas y horas deben ser reformateadas para que Python las interprete como tiempo.

```python
for cada in estaciones:
    cada['TIMESTAMP'] = pd.to_datetime(cada['TIMESTAMP'],infer_datetime_format=True)
```

```python
for channel in canales:
    channel['TIME'] = pd.to_datetime(channel['TIME'],infer_datetime_format=True)
    channel['TIME'] = channel['TIME'] - dt.timedelta(hours=6)
```

```python
for cada in estaciones:
    cada['TEMP_C_Avg'] = (cada['TEMP_C_Avg'] - np.min(cada['TEMP_C_Avg']))/(np.max(cada['TEMP_C_Avg'])-np.min(cada['TEMP_C_Avg']))
```
